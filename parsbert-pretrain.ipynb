```python
from datasets import load_dataset
dataset = load_dataset('text', data_files={'train': [quranpath], 'test':quranpath})
```

```python
def tokenize_pad_and_truncate(texts):
    return tokenizer(texts["text"], padding="max_length", truncation=True)

tokenized_datasets = dataset.map(tokenize_pad_and_truncate, batched=True)
```

```python
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(tokenizer, mlm_probability=0.15)
```

```python
from transformers import  TrainingArguments, Trainer
training_args = TrainingArguments(
#     output_dir="/kaggle/working/Quran_QA3/quranBERT_moreepochs",
    output_dir="/kaggle/working/input/parsbert-pretrain/Quran_QA3/quranBERT_moreepochs",
    save_strategy='epoch',
    evaluation_strategy="epoch",
    learning_rate=1e-5,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=2,
    weight_decay=0.01,
    report_to="none"
)
```

```python
from random import randint
testset =tokenized_datasets['test'].select([randint(0 , 6000) for _ in range(200)])
testset
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset= tokenized_datasets['train'] ,
    eval_dataset = testset,
    data_collator=data_collator,
    tokenizer= tokenizer, 
)
```

```python
trainer.train()
```

```python
os.mkdir("/kaggle/working/Quran_QA3/quranBERT")
os.mkdir("/kaggle/working/Quran_QA3/quranBERT/8epochs")
model.save_pretrained("/kaggle/working/Quran_QA3/quranBERT/8epochs")
tokenizer.save_pretrained("/kaggle/working/Quran_QA3/quranBERT/8epochs")
```
