```python
def dump_jsonl(data, output_path, append=False):
    """
    Write list of objects to a JSON lines file.
    """
    mode = 'a+' if append else 'w'
    with open(output_path, mode, encoding='utf-8') as f:
        for line in data:
            json_record = json.dumps(line, ensure_ascii=False)
            f.write(json_record + '\n')
    print('Wrote {} records to {}'.format(len(data), output_path))
```

```python
with open('/content/drive/MyDrive/farsi_factoid_v2.jsonl', 'r',encoding="utf8") as json_file:
    json_list = list(json_file)

def findOccurrences(s, ch):
    return [i for i, letter in enumerate(s) if letter == ch]

dict_data_list = []

for json_str in json_list:
  result = json.loads(json_str)
  context = result['context']
  context = re.sub("[\(\[].*?[\)\]]", "", context)
  answer = result['answers']
  answer_freq = context.count(answer)

  if answer in context:
    if answer_freq > 1 and "*" in context:
      markers = findOccurrences(context, '*')
      start_char = markers[0]
      context = context.replace('*', '')
    else:
      start_char = context.find(answer)

    answers_list = []
    answer_dict = {}

    answer_dict['text'] = answer
    answer_dict['start_char'] = start_char

    answers_list.append(answer_dict)
    result['answers'] = answers_list
    result['context'] = context

    dict_data_list.append(result)

output_path = '/content/drive/MyDrive/farsi_factoid_v4.jsonl'
dump_jsonl(dict_data_list, output_path)
```
