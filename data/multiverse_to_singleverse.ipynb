```python
with open('/content/drive/MyDrive/total_temp_2700.jsonl', 'r') as f:
  dict_data_list = []
  for line in f:
   data = json.loads(line)
   dict_data_list.append(data)

with open('/content/drive/MyDrive/makarem.xml', 'r', encoding="utf8") as f:
  content = f.read()

soup= BeautifulSoup(content, 'xml')

def findOccurrences(s, ch):
    return [i for i, letter in enumerate(s) if letter == ch]

new_data_list = []

for qa in dict_data_list:
  question = qa['question']
  context = qa['context']
  answer = qa['answers']
  chapter = qa['pq_id'].split(":", 1)[0]
  verses = (qa['pq_id'].split(":", 1)[1]).split('_')[0]
  start_verse = int(verses.split('-')[0])
  end_verse = int(verses.split('-')[1])
  new_data = {}

  if start_verse == end_verse:
    verse_id = 's'+ str(chapter) + '.' + str(start_verse)
    new_data['pq_id'] = verse_id
    new_data['context'] = context
    new_data['answers'] = answer
    new_data['question'] = question
    new_data_list.append(new_data)

  if start_verse != end_verse:
    verse_list = []
    for i in range(start_verse, end_verse+1):
      verse_dict = {}
      verse_id = 's'+ str(chapter) + '.' + str(i)
      verse_content = soup.find(id=verse_id).contents[0]
      verse_dict['id'] = verse_id
      verse_dict['content'] = verse_content
      verse_list.append(verse_dict)

    if context.count(answer) == 1:
      for verse in verse_list:
        if answer in verse['content']:
          new_data['pq_id'] = verse['id']
          new_data['context'] = verse['content']
          new_data['answers'] = answer
          new_data['question'] = question
          new_data_list.append(new_data)
          continue

    if context.count(answer) > 1 and "*" in context:
      markers = findOccurrences(context, '*')
      start_char = markers[0]
      context = context.replace('*', '')
      answer_bounds = context[start_char:start_char+20]
      for verse in verse_list:
        if answer_bounds in verse['content']:
          new_data['pq_id'] = verse['id']
          new_data['context'] = verse['content']
          new_data['answers'] = answer
          new_data['question'] = question
          new_data_list.append(new_data)
          continue

    # else:
    #   exceptions.append(qa)


```

```python
def dump_jsonl(data, output_path, append=False):
    """
    Write list of objects to a JSON lines file.
    """
    mode = 'a+' if append else 'w'
    with open(output_path, mode, encoding='utf-8') as f:
        for line in data:
            json_record = json.dumps(line, ensure_ascii=False)
            f.write(json_record + '\n')
    print('Wrote {} records to {}'.format(len(data), output_path))

output_path = '/content/drive/MyDrive/single_verse.jsonl'
dump_jsonl(new_data_list, output_path)
```
