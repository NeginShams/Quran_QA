```python
import json

with open('C:/Users/Partiran/desktop/QuranInfo.json', 'r', encoding = 'utf_8_sig') as json_file:
    quran_data_list = json.load(json_file)

quran_info = {}

for each in quran_data_list:
    quran_info[each['chapter_name']] = each['chapter_number']
```

```python
from bs4 import BeautifulSoup

with open('C:/Users/Partiran/desktop/makarem.xml', 'r', encoding="utf8") as f:
  content = f.read()

soup= BeautifulSoup(content, 'xml')

exceptions = []
data_list = []

for candidate in final_list:
    spl_word = 'آیه'
    question = candidate['question']
    res = question[question.find(spl_word)+len(spl_word):]
    verse_number = res.split()[0]

    spl_word = 'سوره'
    question = candidate['question']
    res = question[question.find(spl_word)+len(spl_word):]
    chapter_name = res.split()[0]
    chapter_name = chapter_name.replace("،","")

    if chapter_name == '‌ی':
        chapter_name = res.split()[1]

    if chapter_name == 'صاد':
        chapter_name = 'ص'

    if chapter_name == 'قاف':
        chapter_name = 'ق'

    if chapter_name == 'آل‌عمران':
        chapter_name = 'آل عمران'

    if chapter_name == 'ضُحی':
        chapter_name = 'ضحی'

    if chapter_name == 'فصلّت' or chapter_name == 'فصّلت':
        chapter_name = "فصلت"

    if chapter_name == 'نحل\u200cکدام':
        chapter_name = 'نحل'

    if chapter_name == 'حِجر':
        chapter_name = 'حجر'

    if chapter_name == 'ممتحته':
        chapter_name = 'ممتحنه'

    if chapter_name == 'آل':
        chapter_name = "آل عمران"

    if chapter_name == 'انبیا':
        chapter_name = "انبیاء"

    if chapter_name == 'اِسراء':
        chapter_name = "اسراء"

    if chapter_name == 'شرح':
        chapter_name = "انشراح"

    if chapter_name == 'حاقّه':
        chapter_name = "حاقه"
    
    
    

    chapter_name = chapter_name.replace("،","")


    # verse_number = int(verse_number)
    chapter_number = quran_info[chapter_name]

    if verse_number == 'اول':
        verse_number = '1'

    if verse_number == 'دوم':
        verse_number = '2'

    if verse_number == 'سوم':
        verse_number = '3'

    if verse_number == 'هفتم':
        verse_number = '7'

    if verse_number == 'نهم':
        verse_number = '9'

    verse_number = verse_number.replace("سوره","")

    verse_id = 's' + chapter_number + '.' + verse_number

    if verse_id == 's109.آخر':
        verse_id = 's109.6'

    if verse_id == 's106.\u200cی':
        verse_id = 's106.4'

    if verse_id == 's93.آخر':
        verse_id = 's93.11'

    if verse_id == 's48.آخر':
        verse_id = 's48.29'

    if verse_id == 's48.پایانی':
        verse_id = 's48.29'

    if verse_id == 's113.آخر':
        verse_id = 's113.5'

    if verse_id == 's103.اولیه':
        verse_id = 's103.2'

    if verse_id == 's89.آخر':
        verse_id = 's89.30'

    if verse_id == 's103.آخر':
        verse_id = 's103.3'

    if verse_id == 's96.ابتدایی':
        verse_id = 's96.1'

    if verse_id == 's96.':
        verse_id = 's96.1'

    
    
    
    # print(verse_id)

    new_data = {}
    new_data['question'] = candidate['question']
    new_data['answers'] = candidate['possible_answers']
    new_data['pq_id'] = verse_id
    new_data['verse_content'] = soup.find(id=verse_id).contents[0]

    # print(new_data)
    # print(verse_id)

    data_list.append(new_data)


```
