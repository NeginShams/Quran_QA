```python
from bs4 import BeautifulSoup

with open('C:/Users/Partiran/desktop/makarem.xml', 'r', encoding="utf8") as f:
  content = f.read()

soup= BeautifulSoup(content, 'xml')

exceptions = []
data_list = []

for candidate in candidate_list:
    spl_word = 'آیه'
    question = candidate['question']
    res = question[question.find(spl_word)+len(spl_word):]
    verse_number = res.split()[0]
    if verse_number == '‌ی':
        verse_number = res.split()[1]

    verse_number = verse_number.translate(translation_table)

    verse_number = verse_number.replace("،","")

    # if verse_number == '۲۳':
    #     verse_number = '23'

    # if verse_number == '۹۵':
    #     verse_number = '95'

    spl_word = 'سوره'
    question = candidate['question']
    res = question[question.find(spl_word)+len(spl_word):]
    chapter_name = res.split()[0]
    chapter_name = chapter_name.replace("،","")

    chapter_name = chapter_name.strip()

    if chapter_name == '‌ی':
        chapter_name = res.split()[1]

    if chapter_name == 'صاد':
        chapter_name = 'ص'

    if chapter_name == 'قاف':
        chapter_name = 'ق'

    if chapter_name == 'آل‌عمران':
        chapter_name = 'آل عمران'

    if chapter_name == 'ضُحی':
        chapter_name = 'ضحی'

    if chapter_name == 'آل':
        chapter_name = "آل عمران"

    if chapter_name == 'عنكبوت':
        chapter_name = "عنکبوت"

    if chapter_name == 'ملك':
        chapter_name = "ملک"

    if chapter_name == 'شعرا':
        chapter_name = "شعراء"

    if chapter_name == 'فصلّت':
        chapter_name = "فصلت"

    

    if chapter_name == '\u200c':
        chapter_name = res.split()[1]


    chapter_name = chapter_name.replace("،","")

    if len(chapter_name)>3 and "\u200c" in chapter_name and chapter_name!='آل‌عمران':
        # print(chapter_name)
        chapter_name = chapter_name.replace("\u200c","")


    # verse_number = int(verse_number)
    if chapter_name == 'تكویر':
        chapter_number = '81'

    elif chapter_name == 'فصلّت':
        chapter_number = '41'

    elif chapter_name == 'عنكبوت':
        chapter_number = '29'
        
    else:
        chapter_number = quran_info[chapter_name]

    if verse_number == 'اول':
        verse_number = '1'

    if verse_number == 'دوم':
        verse_number = '2'

    if verse_number == 'سوم':
        verse_number = '3'

    

    verse_id = 's' + chapter_number + '.' + verse_number

    # print(verse_id)

    new_data = {}
    new_data['question'] = candidate['question']
    new_data['answers'] = candidate['possible_answers']
    new_data['pq_id'] = verse_id
    new_data['verse_content'] = soup.find(id=verse_id).contents[0]

    # print(new_data)
    # print(verse_id)

    data_list.append(new_data)


```
