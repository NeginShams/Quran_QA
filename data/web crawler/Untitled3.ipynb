```python
import requests
from bs4 import BeautifulSoup
```

```python
home_url = 'https://gharaati.ir/category/%d8%af%d8%b1%d8%b3-%d9%87%d8%a7%db%8c%db%8c-%d8%a7%d8%b2-%d9%82%d8%b1%d8%a2%d9%86/1387/'

page = requests.get(home_url, timeout=100)

soup = BeautifulSoup(page.content, "html.parser")

results = soup.find_all(class_ = "cat-item")

url_list = []

for each in results[1:17]:
    children = each.findChildren()
    for child in children:
        if child.name=='a':
            url_list.append(child.get('href'))
            # print(child.get('href'))
```

```python
for url in url_list:
    print(url)
```

```python
dict_data_list = []
```

```python

for URL in url_list[:2]:
    
        page = requests.get(URL, timeout=100)
        
        soup = BeautifulSoup(page.content, "html.parser")
        
        
        results = soup.find_all(class_ = "post-title post-url")
        
        test_list = ['آیه', 'سوره', 'قرآن']
        
        
        for each in results:
            new_url = each.get('href')
            # print(new_url)
            new_page = requests.get(new_url, timeout=100)
            new_soup = BeautifulSoup(new_page.content, "html.parser")
        
            my_div = new_soup.find(class_ = "entry-content clearfix single-post-content")
            
            children = my_div.findChildren("p" , recursive=False)
            
            for child in children[-5:]:
                new_result = {}
                
                question = (child.text).splitlines()[0]
                possible_answers = (child.text).splitlines()[1:]
        
                res = any(ele in question for ele in test_list)
        
                if res:
        
                    new_result['question'] = question[2:].strip()
                    # new_result['possible_answers'] = possible_answers
                    
                    new_result['possible_answers'] = []
                    for ans in possible_answers:
                        new_result['possible_answers'].append(ans[2:].strip())
                    
                    new_result['url'] = new_url

                    if len(new_result['possible_answers'])==3:
            
                        dict_data_list.append(new_result)

    # except Exception as e:
    #     print(e)
            
```

```python
exceptions = []

for URL in url_list[2:]:
    page = requests.get(URL, timeout=100)
    
    soup = BeautifulSoup(page.content, "html.parser")
    
    
    results = soup.find_all(class_ = "post-title post-url")
    
    test_list = ['آیه', 'سوره', 'قرآن']
    
    
    for each in results:
        try:
            new_url = each.get('href')
            # print(new_url)
            new_page = requests.get(new_url, timeout=100)
            new_soup = BeautifulSoup(new_page.content, "html.parser")
        
            my_div = new_soup.find(class_ = "entry-content clearfix single-post-content")
    
            children = my_div.findChildren("p")
                
            para = children[-1]
            
            if len(para.text) <10:
                para = children[-2]
    
            # print(para)
            # print('@@@@@@@@@@@@@@@@@@@@@@@@@@@222222222')
                
                
            i = 0
                
            while i!= 20:
                new_data = {}
                try:
            
            # q = para[0].text.splitlines()[i]
                    q = para.text.splitlines()[i]
                    
                    res = any(ele in q for ele in test_list)
                
                    if res:
                        new_data['question'] = q[2:].strip()
                        new_data['possible_answers'] = []
                        for j in range(i+1, i+4):
                            # new_data['answers'].append(para[0].text.splitlines()[j])
                            answer = para.text.splitlines()[j]
                            new_data['possible_answers'].append(answer[2:].strip())
        
                        new_data['url'] = new_url
                    
                        dict_data_list.append(new_data)
                except:
                    exceptions.append(new_url)
                    
                i = i+4
        except:
            exceptions.append(new_url)
        
        # i = i+4
```
