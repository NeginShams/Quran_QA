```python
# url = 'https://gharaati.ir/%d8%a7%d8%b3%d8%aa%d8%a7%d9%86%d8%af%d8%a7%d8%b1%d8%af-%d8%af%d8%b1-%d8%a2%db%8c%d8%a7%d8%aa-%d9%88-%d8%b1%d9%88%d8%a7%db%8c%d8%a7%d8%aa-1/'
url = 'https://gharaati.ir/%d8%af%d9%81%d8%a7%d8%b9-%d8%a7%d8%b2-%da%a9%d8%b4%d9%88%d8%b1-%d8%af%d8%b1-%d8%a8%d8%b1%d8%a7%d8%a8%d8%b1-%d8%af%d8%b4%d9%85%d9%86/'


page = requests.get(url, timeout=100)

soup = BeautifulSoup(page.content, "html.parser")

my_div = soup.find(class_ = "entry-content clearfix single-post-content")

children = my_div.findChildren("p" , recursive=False)

ele = children[-1:]

i = 0
qa_list = []

while i!= 20:
    new_data = {}
    q = ele[0].text.splitlines()[i]
    new_data['question'] = q[2:].strip()
    new_data['answers'] = []
    for j in range(i+1, i+4):
        new_data['answers'].append(ele[0].text.splitlines()[j])

    qa_list.append(new_data)

    i = i+4

for qa in qa_list:
    print(qa)
    print('!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!')
```
