```python
from bs4 import BeautifulSoup
from deep_translator import GoogleTranslator
translator = GoogleTranslator(source='ar',target='fa')

with open('/content/drive/MyDrive/makarem.xml', 'r', encoding="utf8") as f:
    content = f.read()

soup= BeautifulSoup(content, 'xml')

not_complete_list = []

complete_list = []

multiple_ocuurence = []

for qa in factoid_dict_data:
  farsi_question = translator.translate(qa['Question_Text'])
  farsi_answer_instance = translator.translate(qa['Answer-Instances '].strip())

  if type(qa['Expert_Commentary']) == float and math.isnan(qa['Expert_Commentary']):
    farsi_answer_expert = ''
  else:
    clean_answer = qa['Expert_Commentary'].replace("ﷺ","")
    clean_answer = clean_answer.replace("❂","")
    clean_answer = clean_answer.strip()
    # farsi_answer_expert = translator.translate(qa['Expert_Commentary'])
    farsi_answer_expert = translator.translate(clean_answer)

  chapter = qa['Chapter_Number']
  start_verse = qa['Verses_Number_Start']
  end_verse = qa['Verses_Number_End']

  pq_id = str(chapter) + ':'+ str(start_verse) + '-' + str(end_verse)

  farsi_passage = ''
  for i in range(start_verse, end_verse+1):
    verse_id = 's'+ str(chapter) + '.' + str(i)
    translated_verse = soup.find(id=verse_id).contents[0]
    farsi_passage += translated_verse
    farsi_passage += ' '

  new_data = {}
  new_data['context'] = farsi_passage
  new_data['question'] = farsi_question
  new_data['pq_id'] = pq_id

  if farsi_answer_instance in farsi_passage:
    new_data['answer'] = farsi_answer_instance

    if farsi_passage.count(farsi_answer_instance) == 1:
      complete_list.append(new_data)
    else:
      multiple_ocuurence.append(new_data)


  elif farsi_answer_expert and farsi_answer_expert in farsi_passage:
    new_data['answer'] = farsi_answer_expert
    if farsi_passage.count(farsi_answer_expert) == 1:
      complete_list.append(new_data)
    else:
      multiple_ocuurence.append(new_data)

  else:
    new_data['answer'] = qa['Answer-Instances '].strip() + ' (' + farsi_answer_instance + ') ' + '(' + farsi_answer_expert + ')'
    not_complete_list.append(new_data)
```

```python
import json

def dump_jsonl(data, output_path, append=False):
    """
    Write list of objects to a JSON lines file.
    """
    mode = 'a+' if append else 'w'
    with open(output_path, mode, encoding='utf-8') as f:
        for line in data:
            json_record = json.dumps(line, ensure_ascii=False)
            f.write(json_record + '\n')
    print('Wrote {} records to {}'.format(len(data), output_path))

```

```python
complete_output_path = '/content/drive/MyDrive/QUQA_translated/QUQA_complete.jsonl'
not_complete_path = '/content/drive/MyDrive/QUQA_translated/QUQA_not_complete.jsonl'
multiple_ocuurence_path = '/content/drive/MyDrive/QUQA_translated/QUQA_multiple_ocuurence.jsonl'

dump_jsonl(complete_list, complete_output_path)
dump_jsonl(not_complete_list, not_complete_path)
dump_jsonl(multiple_ocuurence, multiple_ocuurence_path)
```
