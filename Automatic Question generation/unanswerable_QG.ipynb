```python
factoid_QA = df[df["Quetion_Type"] == 'F']
```

```python
import json
from deep_translator import GoogleTranslator
translator = GoogleTranslator(source='ar',target='fa')

qa_list = []

for qa in factoid_dict_data:
  farsi_question = translator.translate(qa['Question_Text'])
  farsi_answer_instance = translator.translate(qa['Answer-Instances '].strip())

  new_data = {}
  new_data['question'] = farsi_question
  new_data['answer'] = farsi_answer_instance

  qa_list.append(new_data)

# with open('/content/drive/MyDrive/QUQA_translated/haqa_factoid.json', 'w') as fp:
#   json.dump(qa_list, fp, ensure_ascii = False)
```

```python
from pyserini.search.lucene import LuceneSearcher
from bs4 import BeautifulSoup

with open('/content/drive/MyDrive/makarem.xml', 'r', encoding="utf8") as f:
  content = f.read()

soup= BeautifulSoup(content, 'xml')

for qa in qa_list:
  searcher = LuceneSearcher('/content/drive/MyDrive/indexes/sample_collection_jsonl')
  searcher.set_language('fa')
  hits = searcher.search(qa['question'])
  verse_id = hits[0].docid
  qa['pq_id'] = verse_id
  qa['context'] = soup.find(id=verse_id).contents[0]
```
